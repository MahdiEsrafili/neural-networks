```python
def conv(inpt, fltr):
    fmap = np.zeros((inpt.shape[0]-fltr.shape[1]+2,
                    inpt.shape[1] - fltr.shape[2]+2,
                    flt.shape[0]))
    for f in range(fltr.shapep[0]):
        fmap[:,:,f] = conv_(inpt, fltr[f,:,:])
        
    return fmap
```

```python
def conv_(f1,f2):
    
    #add zero padding
    f1 = np.pad(f1, ((f2.shape[0]-1,f2.shape[0]-1),(f2.shape[1]-1,f2.shape[1]-1)))
    res = np.zeros((f1.shape[0] - f2.shape[0]+2,f1.shape[1] - f2.shape[1]+2))
    for r in range(f1.shape[0] - f2.shape[0]+1):
        for c in range(f1.shape[1] - f2.shape[1]+1):
            #print(r,c)
            temp_res = np.sum(f1[r:r+f2.shape[0],c:c+f2.shape[1]] * f2)
            res[r,c] = temp_res
            #print(res)
    #res = res[f2.shape[0]-1:res.shape[0]-f2.shape[0],f2.shape[1]-1:res.shape[1]-f2.shape[1]]
    return res
```

```python
def relu(inpt):
    return np.maximum(inpt,0)
```

```python
def max_pooling(inpt):
    stride=2
    ksize = 2
    h = int((inpt.shape[0] - ksize)/stride +1)
    w = int((inpt.shape[1] - ksize)/stride +1)
    res = np.empty((h,w,inpt.shape[2]))
    print(res.shape)
    for l in range(inpt.shape[2]):
        for i in range(res.shape[0]):
            for j in range(res.shape[1]):
                res[i, j, l] = np.max(inpt[i*stride:i*stride+ksize,j*stride:j*stride+ksize,l])
                
    return res
            
        
```
